```python
import pandas as pd
import ast
import re
from pathlib import Path

# Pad naar je CSV
CSV_PATH = "country_mentions_raw_good.csv"

# KIES HIER JE TEKSTBESTAND
# Voorbeeld: "TXT/Session 01 - 1946.txt"
TEXT_FILE = Path("TXT/Session 01 - 1946/AUS_01_1946.txt")

WINDOW = 3  # aantal woorden vóór en ná het targetwoord
# KIES HIER WELK LAND JE WILT (als speaker)
SPEAKER_FILTER = "AUS"   # bijv. "AUS" of "ARG" of None

# Of als je liever op target_iso filtert:
TARGET_FILTER = None 

def load_text():
    with open(TEXT_FILE, encoding="utf-8") as f:
        return f.read()

def build_word_index(text):
    words = []
    for m in re.finditer(r"\S+", text):
        words.append({
            "word": m.group(),
            "start": m.start(),
            "end": m.end()
        })
    return words

def get_context(words, abs_start, abs_end, window=3):
    idx = None
    for i, w in enumerate(words):
        if (w["start"] <= abs_start < w["end"]) or (w["start"] < abs_end <= w["end"]):
            idx = i
            break
    if idx is None:
        idx = min(
            range(len(words)),
            key=lambda i: abs(words[i]["start"] - abs_start)
        )
    start_idx = max(0, idx - window)
    end_idx = min(len(words), idx + window + 1)
    context_words = [w["word"] for w in words[start_idx:end_idx]]
    return " ".join(context_words)

def main():
    # alleen de eerste 10 rijen inlezen
    df = pd.read_csv(CSV_PATH, nrows=30)

    # >>> HIER FILTEREN OP LAND <<<
    if SPEAKER_FILTER is not None:
        df = df[df["speaker_iso"] == SPEAKER_FILTER]

    if TARGET_FILTER is not None:
        df = df[df["target_iso"] == TARGET_FILTER]

    # Als er na filteren niets overblijft:
    if df.empty:
        print("Geen rijen gevonden met deze filter(s).")
        return

    text = load_text()
    words = build_word_index(text)

    for _, row in df.iterrows():
        positions = ast.literal_eval(row["positions"])

        for pos in positions:
            abs_start = pos["abs_start"]
            abs_end = pos["abs_end"]
            surface = pos["surface"]

            context = get_context(words, abs_start, abs_end, window=WINDOW)

            print(
                f"{row['speaker_iso']},{row['target_iso']},"
                f"{row['year']},{row['session']},"
                f"surface='{surface}' -> {context}"
            )

if __name__ == "__main__":
    main()
```

```python
import os
import glob
import dash
from dash import dcc, html, Input, Output, State
import plotly.express as px
import pandas as pd

# ─────────────────────────────────────────────
# CONFIGURATIE
# ─────────────────────────────────────────────
DATA_ROOT = "TXT"          # map waar de Session-mappen staan
# Pas dit aan als de mappen ergens anders staan, bijv.:
# DATA_ROOT = "C:/Users/jij/speeches"

# ─────────────────────────────────────────────
# HULPFUNCTIES
# ─────────────────────────────────────────────

def get_sessions() -> dict[int, str]:
    """Geeft {jaar: mappad} voor alle gevonden Session-mappen."""
    sessions = {}
    for folder in sorted(glob.glob(os.path.join(DATA_ROOT, "Session * - *"))):
        basename = os.path.basename(folder)
        try:
            year = int(basename.split(" - ")[-1].strip())
            sessions[year] = folder
        except ValueError:
            pass
    return sessions


def get_countries_for_session(folder: str) -> dict[str, str]:
    """Geeft {ISO3-code: bestandspad} voor alle txt-bestanden in een sessie-map."""
    countries = {}
    for filepath in glob.glob(os.path.join(folder, "*.txt")):
        filename = os.path.basename(filepath)
        iso3 = filename.split("_")[0].upper()
        countries[iso3] = filepath
    return countries


def read_speech(filepath: str, max_chars: int = 3000) -> str:
    """Leest een tekstbestand en geeft de inhoud terug (afgekapt op max_chars)."""
    try:
        with open(filepath, "r", encoding="utf-8", errors="replace") as f:
            text = f.read(max_chars)
        if os.path.getsize(filepath) > max_chars:
            text += "\n\n[... tekst afgekapt ...]"
        return text
    except Exception as e:
        return f"Fout bij laden: {e}"


# ─────────────────────────────────────────────
# DATA LADEN
# ─────────────────────────────────────────────
sessions = get_sessions()
years = sorted(sessions.keys())

if not years:
    raise FileNotFoundError(
        f"Geen Session-mappen gevonden in '{DATA_ROOT}'. "
        "Pas DATA_ROOT aan naar de juiste locatie."
    )

# ─────────────────────────────────────────────
# DASH APP
# ─────────────────────────────────────────────
app = dash.Dash(__name__)
app.title = "UN Speeches – Interactieve Kaart"

app.layout = html.Div([

    html.H2("🌍 UN General Debate Speeches",
            style={"textAlign": "center", "fontFamily": "Arial", "marginBottom": "5px"}),

    # Jaar-slider
    html.Div([
        dcc.Slider(
            id="year-slider",
            min=min(years),
            max=max(years),
            step=1,
            value=min(years),
            marks={y: str(y) for y in years if y % 5 == 0 or y == min(years) or y == max(years)},
            tooltip={"placement": "bottom", "always_visible": True},
        )
    ], style={"width": "90%", "margin": "0 auto 10px auto"}),

    # Kaart
    dcc.Graph(id="world-map", style={"height": "65vh"}),

    # Popup / speech-venster
    html.Div(id="speech-panel", children=[
        html.Div("Klik op een land om de speech te lezen.",
                 style={"color": "#888", "fontStyle": "italic"})
    ], style={
        "width": "90%",
        "margin": "10px auto",
        "padding": "15px",
        "border": "1px solid #ccc",
        "borderRadius": "8px",
        "maxHeight": "300px",
        "overflowY": "auto",
        "fontFamily": "monospace",
        "fontSize": "13px",
        "backgroundColor": "#f9f9f9",
        "whiteSpace": "pre-wrap",
    }),

], style={"maxWidth": "1200px", "margin": "0 auto"})


# ─────────────────────────────────────────────
# CALLBACKS
# ─────────────────────────────────────────────

@app.callback(
    Output("world-map", "figure"),
    Input("year-slider", "value")
)
def update_map(year):
    folder = sessions[year]
    countries = get_countries_for_session(folder)
    iso_list = list(countries.keys())

    df = pd.DataFrame({
        "iso_alpha": iso_list,
        "Beschikbaar": [1] * len(iso_list)
    })

    fig = px.choropleth(
        df,
        locations="iso_alpha",
        color="Beschikbaar",
        color_continuous_scale=[[0, "#b3d9ff"], [1, "#005f99"]],
        projection="natural earth",
        title=f"Sessie {year} — {len(iso_list)} landen met speech",
    )
    fig.update_coloraxes(showscale=False)
    fig.update_layout(
        margin={"r": 0, "t": 40, "l": 0, "b": 0},
        geo=dict(showframe=False, showcoastlines=True),
        clickmode="event+select",
    )
    return fig


@app.callback(
    Output("speech-panel", "children"),
    Input("world-map", "clickData"),
    State("year-slider", "value"),
    prevent_initial_call=True
)
def show_speech(clickData, year):
    if clickData is None:
        return "Klik op een land om de speech te lezen."

    iso3 = clickData["points"][0].get("location", "")
    folder = sessions[year]
    countries = get_countries_for_session(folder)

    if iso3 not in countries:
        return f"Geen speech gevonden voor {iso3} in {year}."

    text = read_speech(countries[iso3])
    return [
        html.Strong(f"🗣 {iso3} — {year}", style={"fontSize": "15px"}),
        html.Hr(),
        html.Pre(text, style={"whiteSpace": "pre-wrap", "fontFamily": "monospace"})
    ]


# ─────────────────────────────────────────────
if __name__ == "__main__":
    app.run(debug=True)
```

```python
# app.py

import os
import re
import json
import pandas as pd
import geopandas as gpd
from dash import Dash, dcc, html, Input, Output, State
import plotly.express as px

# -----------------------------
# 1. SPEECHES INLADEN
# Verwachte mapstructuur:
#   Session 01 - 1946/ARG_01_1946.txt
#   Session 02 - 1947/USA_02_1947.txt
# -----------------------------
SPEECHES_ROOT = "."  # pas aan als de mappen ergens anders staan

def load_speeches(root: str) -> dict:
    """
    Geeft een dict terug: {(iso3, year): tekst}
    """
    speeches = {}
    session_pattern = re.compile(r"Session \d+ - (\d{4})", re.IGNORECASE)

    for folder in os.listdir(root):
        folder_path = os.path.join(root, folder)
        if not os.path.isdir(folder_path):
            continue
        m = session_pattern.search(folder)
        if not m:
            continue
        year = int(m.group(1))

        for fname in os.listdir(folder_path):
            if not fname.endswith(".txt"):
                continue
            iso3 = fname[:3].upper()
            fpath = os.path.join(folder_path, fname)
            try:
                with open(fpath, "r", encoding="utf-8", errors="replace") as f:
                    speeches[(iso3, year)] = f.read()
            except Exception:
                pass
    return speeches

speeches = load_speeches(SPEECHES_ROOT)

# -----------------------------
# 2. MENTIONS DATA INLADEN
# -----------------------------
df = pd.read_csv("country_mentions_raw_good.csv")

mentions = (
    df.groupby(["speaker_iso", "target_iso", "year"], as_index=False)["count"]
      .sum()
      .rename(columns={"count": "mentions"})
)

code_map = {
    "URS": "SUN",
    "GIN": "PNG",
}
mentions["target_iso_mapped"] = mentions["target_iso"].replace(code_map)

# -----------------------------
# 3. CSHAPES INLADEN EN PER JAAR UITVOUWEN
# -----------------------------
shp_path = "cshapes-2.0.shp"
gdf = gpd.read_file(shp_path)

iso_col  = "ISO1AL3"
to_col   = "GWEYEAR"
from_col = "GWSYEAR" if "GWSYEAR" in gdf.columns else "COWSYEAR"

MIN_YEAR = 1945
MAX_YEAR = 2019

records = []
for _, row in gdf.iterrows():
    start = max(int(row[from_col]), MIN_YEAR)
    end   = min(int(row[to_col]),   MAX_YEAR)
    for y in range(start, end + 1):
        new_row = row.copy()
        new_row["year"] = y
        records.append(new_row)

gdf_yearly = gpd.GeoDataFrame(records, crs=gdf.crs)
gdf_yearly = gdf_yearly.to_crs(epsg=4326)

# -----------------------------
# 4. DASH-APP
# -----------------------------
app = Dash(__name__)

app.layout = html.Div([
    html.H2("VN-speeches: wie noemt wie?"),

    # --- Controls ---
    html.Div([
        html.Div([
            html.Label("Spreker (speaker_iso):"),
            dcc.Dropdown(
                id="speaker-dropdown",
                options=[{"label": s, "value": s}
                         for s in sorted(mentions["speaker_iso"].unique())],
                value="ARG",
                clearable=False
            ),
        ], style={"width": "30%", "display": "inline-block"}),

        html.Div([
            html.Label("Jaar:"),
            dcc.Slider(
                id="year-slider",
                min=int(mentions["year"].min()),
                max=int(mentions["year"].max()),
                value=int(mentions["year"].min()),
                marks={int(y): str(y) for y in range(
                    int(mentions["year"].min()),
                    int(mentions["year"].max()) + 1, 5)},
                step=1
            ),
        ], style={"width": "60%", "display": "inline-block", "padding": "0 20px"}),
    ]),

    # --- Kaart ---
    dcc.Graph(id="world-map", style={"height": "600px"}),

    # --- Speech popup ---
    html.Div(
        id="speech-modal",
        children=[
            html.Div([
                html.Div([
                    html.H3(id="speech-title", style={"display": "inline-block", "marginRight": "20px"}),
                    html.Button("✕ Sluiten", id="close-modal", n_clicks=0,
                                style={"float": "right", "cursor": "pointer",
                                       "fontSize": "14px", "padding": "4px 10px"}),
                ]),
                html.Hr(),
                html.Pre(
                    id="speech-text",
                    style={
                        "whiteSpace": "pre-wrap",
                        "maxHeight": "400px",
                        "overflowY": "auto",
                        "fontFamily": "Georgia, serif",
                        "fontSize": "13px",
                        "lineHeight": "1.6",
                    }
                ),
            ], style={
                "background": "white",
                "padding": "20px",
                "borderRadius": "8px",
                "maxWidth": "700px",
                "margin": "40px auto",
                "boxShadow": "0 4px 20px rgba(0,0,0,0.3)",
                "position": "relative",
            }),
        ],
        style={"display": "none",   # verborgen totdat er geklikt wordt
               "position": "fixed",
               "top": 0, "left": 0,
               "width": "100%", "height": "100%",
               "background": "rgba(0,0,0,0.5)",
               "zIndex": 9999,
               "overflowY": "auto"},
    ),
])


# -----------------------------
# 5. CALLBACKS
# -----------------------------

@app.callback(
    Output("world-map", "figure"),
    Input("speaker-dropdown", "value"),
    Input("year-slider", "value"),
)
def update_map(selected_speaker, selected_year):
    m_year = mentions[
        (mentions["speaker_iso"] == selected_speaker) &
        (mentions["year"] == selected_year)
    ].copy()
    m_year["target_iso_mapped"] = m_year["target_iso_mapped"].fillna(m_year["target_iso"])

    g_year = gdf_yearly[gdf_yearly["year"] == selected_year].copy()

    merged = g_year.merge(
        m_year,
        left_on=iso_col,
        right_on="target_iso_mapped",
        how="left"
    )
    merged["mentions"] = merged["mentions"].fillna(0)
    merged["id"] = merged[iso_col]

    # Voeg een kolom toe die aangeeft of er een speech beschikbaar is
    merged["heeft_speech"] = merged[iso_col].apply(
        lambda iso: "📄 Speech beschikbaar" if (iso, selected_year) in speeches
        else "Geen speech"
    )

    geo = json.loads(merged[["id", "geometry"]].to_json())

    max_mentions = merged["mentions"].max()
    if max_mentions == 0:
        max_mentions = 1

    fig = px.choropleth(
        merged,
        geojson=geo,
        locations="id",
        featureidkey="properties.id",
        color="mentions",
        color_continuous_scale="Reds",
        range_color=(0, max_mentions),
        hover_data={"id": True, "mentions": True, "heeft_speech": True},
        labels={"mentions": "Aantal mentions", "id": "Land (ISO3)", "heeft_speech": "Speech"},
        title=f"{selected_speaker} → andere landen, jaar {selected_year}",
    )

    fig.update_geos(fitbounds="locations", visible=False)
    fig.update_layout(
        margin={"r": 0, "t": 40, "l": 0, "b": 0},
        coloraxis_colorbar_title="Mentions",
        clickmode="event+select",
    )
    return fig


@app.callback(
    Output("speech-modal", "style"),
    Output("speech-title", "children"),
    Output("speech-text", "children"),
    Input("world-map", "clickData"),
    Input("close-modal", "n_clicks"),
    State("year-slider", "value"),
    prevent_initial_call=True,
)
def toggle_speech(click_data, close_clicks, selected_year):
    """Opent de popup bij klik op een land; sluit bij ✕."""
    from dash import ctx

    modal_hidden = {"display": "none"}
    modal_visible = {
        "display": "block",
        "position": "fixed",
        "top": 0, "left": 0,
        "width": "100%", "height": "100%",
        "background": "rgba(0,0,0,0.5)",
        "zIndex": 9999,
        "overflowY": "auto",
    }

    # Sluit-knop ingedrukt
    if ctx.triggered_id == "close-modal":
        return modal_hidden, "", ""

    # Klik op de kaart
    if click_data is None:
        return modal_hidden, "", ""

    try:
        iso3 = click_data["points"][0]["location"]
    except (KeyError, IndexError):
        return modal_hidden, "", ""

    key = (iso3, selected_year)
    if key in speeches:
        title = f"🗣️ Speech: {iso3} — {selected_year}"
        text  = speeches[key]
    else:
        title = f"{iso3} — {selected_year}"
        text  = "Geen speechbestand gevonden voor dit land en jaar."

    return modal_visible, title, text


if __name__ == "__main__":
    app.run(debug=True)
```
